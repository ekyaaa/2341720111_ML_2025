
Tugas 1: Modifikasi Model CNN CIFAR-10

Modifikasi model CNN pada praktikum 2 sehingga didapatkan akurasi testing lebih dari 80%.

Tips untuk meningkatkan akurasi:

    - Tambahkan lebih banyak layer konvolusi
    - Gunakan Dropout untuk mencegah overfitting
    - Tambahkan Batch Normalization
    - Tingkatkan jumlah epochs
    - Gunakan Data Augmentation
    - Eksperimen dengan learning rate dan optimizer yang berbeda



```python
import tensorflow as tf
from tensorflow.keras import datasets, layers, models, regularizers, optimizers
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
import matplotlib.pyplot as plt
import numpy as np

# 1. Load dan Preprocess Data
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalisasi (Tetap sama karena ini standar)
train_images = train_images.astype('float32') / 255.0
test_images = test_images.astype('float32') / 255.0

# 2. Arsitektur Model (Dimodifikasi: ELU + GlobalAvgPool)
model = models.Sequential([
    # Block 1
    layers.Conv2D(32, (3, 3), padding='same', input_shape=(32, 32, 3)),
    layers.BatchNormalization(),
    layers.Activation('elu'), # Menggunakan ELU
    layers.Conv2D(32, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('elu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.2),

    # Block 2
    layers.Conv2D(64, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('elu'),
    layers.Conv2D(64, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('elu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.3),

    # Block 3
    layers.Conv2D(128, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('elu'),
    layers.Conv2D(128, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('elu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.4),

    # Block 4 (Tambahan layer agar lebih deep dari temanmu)
    layers.Conv2D(256, (3, 3), padding='same'),
    layers.BatchNormalization(),
    layers.Activation('elu'),
    layers.MaxPooling2D((2, 2)),
    layers.Dropout(0.4),

    # Classification Head (BEDA: Pakai Global Average Pooling)
    layers.GlobalAveragePooling2D(),
    layers.Dense(256, kernel_regularizer=regularizers.l2(0.001)),
    layers.BatchNormalization(),
    layers.Activation('elu'),
    layers.Dropout(0.5),
    layers.Dense(10, activation='softmax')
])

model.summary()

# 3. Compile dengan Callbacks (BEDA: Auto-adjust Learning Rate)
opt = optimizers.Adam(learning_rate=0.001)

model.compile(optimizer=opt,
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Callback: Turunkan LR jika val_loss tidak membaik selama 3 epoch
rlr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1, min_lr=0.00001)
# Callback: Stop training jika tidak ada perkembangan berarti (hemat waktu)
es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# 4. Data Augmentation
from tensorflow.keras.preprocessing.image import ImageDataGenerator
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True,
    zoom_range=0.1
)
datagen.fit(train_images)

# 5. Training
history = model.fit(datagen.flow(train_images, train_labels, batch_size=64),
                    epochs=40, # 40 epoch cukup dengan LR scheduler
                    validation_data=(test_images, test_labels),
                    callbacks=[rlr, es],
                    verbose=1)

# 6. Evaluasi Plot
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label = 'Validation Accuracy')
plt.title('Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label = 'Validation Loss')
plt.title('Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.legend(loc='upper right')
plt.show()

test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print(f'\nTest accuracy CIFAR-10: {test_acc*100:.2f}%')
```


Tugas 2: Klasifikasi Dataset MNIST dengan CNN

Buatlah model CNN untuk klasifikasi dataset MNIST.

Petunjuk:

    - Dataset MNIST berisi gambar tulisan tangan angka 0-9
    - Ukuran gambar: 28x28 piksel, grayscale (1 channel)
    - Jumlah kelas: 10 (angka 0-9)
    - Gunakan arsitektur CNN yang sesuai untuk dataset ini



```python
# 1. Load Data
(train_images_mnist, train_labels_mnist), (test_images_mnist, test_labels_mnist) = datasets.mnist.load_data()

# Reshape & Normalize
train_images_mnist = train_images_mnist.reshape((60000, 28, 28, 1)).astype('float32') / 255.0
test_images_mnist = test_images_mnist.reshape((10000, 28, 28, 1)).astype('float32') / 255.0

# 2. Arsitektur Model (Lebih "Padat" dari temanmu)
model_mnist = models.Sequential([
    # Layer 1
    layers.Conv2D(32, (3, 3), padding='same', input_shape=(28, 28, 1)),
    layers.BatchNormalization(), # Penambahan Batch Norm
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    # Layer 2
    layers.Conv2D(64, (3, 3), padding='same'),
    layers.BatchNormalization(), # Penambahan Batch Norm
    layers.Activation('relu'),
    layers.MaxPooling2D((2, 2)),

    # Layer 3
    layers.Conv2D(128, (3, 3), padding='valid'), # Valid padding
    layers.BatchNormalization(),
    layers.Activation('relu'),

    # Dense Layers
    layers.Flatten(),
    layers.Dense(128),
    layers.BatchNormalization(),
    layers.Activation('relu'),
    layers.Dropout(0.5), # Dropout 50%
    layers.Dense(10, activation='softmax')
])

model_mnist.summary()

# 3. Compile
model_mnist.compile(optimizer='adam',
                    loss='sparse_categorical_crossentropy',
                    metrics=['accuracy'])

# 4. Training dengan Sedikit Augmentasi (Opsional tapi bagus untuk nilai plus)
datagen_mnist = ImageDataGenerator(
    rotation_range=10,  # Rotasi sedikit karena angka tulisan tangan bisa miring
    zoom_range=0.1,
    width_shift_range=0.1,
    height_shift_range=0.1
)
datagen_mnist.fit(train_images_mnist)

history_mnist = model_mnist.fit(datagen_mnist.flow(train_images_mnist, train_labels_mnist, batch_size=64),
                                epochs=15, # Tambah sedikit epoch karena ada augmentasi
                                validation_data=(test_images_mnist, test_labels_mnist),
                                verbose=1)

# 5. Evaluasi
plt.plot(history_mnist.history['accuracy'], label='accuracy')
plt.plot(history_mnist.history['val_accuracy'], label = 'val_accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.legend(loc='lower right')
plt.show()

test_loss_mnist, test_acc_mnist = model_mnist.evaluate(test_images_mnist, test_labels_mnist, verbose=2)
print(f'\nTest accuracy MNIST: {test_acc_mnist*100:.2f}%')
```
